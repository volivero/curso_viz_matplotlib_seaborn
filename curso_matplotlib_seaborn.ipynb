# **Subplots**

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2)

axes[1,0].plot(x,np.sinh(x))
axes[1,1].plot(x,np.cosh(y),'y')
axes[1,2].plot(x,np.sinh(y),'r')
axes[1,3].plot(x,np.arcsinh(x))

fig.tight_layout()
```

```python
# Graficando series de tiempo

climate_change = pd.read_csv("/content/final_project_dataset/climate_change.csv", parse_dates=["date"],
                              index_col="date")

sixties = climate_change["1960-01-01":"1969-12-31"] # NUEVO

fig, ax = plt.subplots()

ax.plot(seattle_weather.index, seattle_weather["MLY-TAVG-NORMAL"]) # antiguo
ax.plot(sixties.index, sixties['co2']) # NUEVO

ax.set_xlabel("Times (Months)")
ax.set_ylabel("Average temp (F degress)")
ax.set_title("Weather in Seattle")

plt.show()
```

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', colors=color)
```

```python
# PARA AGREGAR DIFERENTES ESCALAS

fig, ax = plt.subplots()

plot_timeseries(ax, climate_change.index, climate_change['co2'],'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax, climate_change.index,climate_change['relative_temp'],'red', 'Time','Relative temperature (Celsius)')

#Agregar anotaciones

ax2.annotate(">1 degree", xy=(pd.Timestamp('2015-10-06'), 1),
              xytext=(pd.Timestamp('2008-10-06'), -0.2),
              arrowprops={"arrowstyle":"->", "color":"gray"})

plt.show()
```

```python
#Comparacion cuantativa por Scatter Plot

eighties = climate_change["1980-01-01":"1989-12-31"]
nineties = climate_change["1990-01-01":"1999-12-31"]

fig, ax = plt.subplots()
ax.scatter(eighties["co2"], eighties["relative_temp"],
           color="red", label="eighties")
ax.scatter(nineties["co2"], nineties["relative_temp"],
           color="blue", label="nineties")

ax.legend()
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperature (Celsius)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(climate_change["co2"], climate_change["relative_temp"],
           c=climate_change.index)
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperature (Celsius)")
plt.show()
```
